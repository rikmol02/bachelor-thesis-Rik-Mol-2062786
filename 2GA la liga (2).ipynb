```python
start_time = time.time()

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC

import numpy as np
import matplotlib.pyplot as plt
import mglearn

#Splitting data into train and test: merge first 9 seasons
trainData=pd.concat(finalDfList[:8])
trainValData=pd.concat(finalDfList[:9])
valData=finalDfList[8]
testData=finalDfList[9]

#Make sure the outcome of the match is used as the predicted variable
X_train=trainData.iloc[:,1:]
y_train=trainData.iloc[:,0]
X_valid=valData.iloc[:,1:]
y_valid=valData.iloc[:,0]
X_test=testData.iloc[:,1:]
y_test=testData.iloc[:,0]

X_trainval=trainValData.iloc[:,1:]
y_trainval=trainValData.iloc[:,0]

print("Size of training set: {} size of validation set: {} size of test set:"" {}\n".format(X_train.shape[0], X_valid.shape[0], X_test.shape[0]))
best_score = 0
for gamma in [0.000000000000001, 0.000000001, 0.00000001, 0.0000001, 0.000001, 0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100]:
    for C in [0.001, 0.01, 0.1, 1, 10, 100]:
        # for each combination of parameters, train an support vector classifier 
        svm = SVC(gamma=gamma, C=C)
        svm.fit(X_train, y_train)

        # evaluate the SVC on the test set
        score = svm.score(X_valid, y_valid)
        print(f'Score: {round(score,4)}, gamma: {gamma}, C: {C}')
            # if we got a better score, store the score and parameters
        if score > best_score:
            best_score = score
            best_parameters = {'C': C, 'gamma': gamma}
            

# rebuild a model on the combined training and validation set,
# and evaluate it on the test set 
svm = SVC(**best_parameters)
svm.fit(X_trainval, y_trainval)
test_score = svm.score(X_test, y_test)
        
print("Best score on validation set: {:.2f}".format(best_score))
print("Best parameters: ", best_parameters)
print("Test set score with best parameters: {:.2f}".format(test_score)) 
svm_pred=svm.predict(X_test)
print(svm_pred)

print("--- %s seconds ---" % (time.time() - start_time))
```
